```python
class BankAccount:
  def set_value(self,owner_name,balance,account_number):
      self.owner_name = owner_name
      self.balance = balance
      self.account_number = account_number

  def display(self):
    print(f"Account Owner : {self.owner_name} | Account Number : {self.account_number} | Account Balance : {self.balance}")

  def deposit(self, amount):
    if amount > 0:
      self.balance += amount
      print(f"{amount} has beeen deposit to {self.account_number}  name : {self.owner_name}")
    else:
      print("Deposit amount must be positve number")

  def withdraw(self, amount):
    if amount > self.balance:
      print(f" Amount want to withdraw {amount} \n Please enter an amount which is equal or less than {self.balance} ")

    elif amount > 0 and amount <= self.balance :
      self.balance -= amount
      print(f"The amount {amount} has been withdrawed !\nCurrent balance {self.balance}")



  def check_balance(self):
    if self.balance < 0 :
      print(f"{self.owner_name} : The account is empty")
    else:
      print(f"{self.owner_name} : The account has {self.balance}")

a1 = BankAccount()
a2 = BankAccount()
a3 = BankAccount()

a1.set_value ('Abhi',2000,'AT1000000')
a2.set_value ('Divya',3000,'AT1000001')
a3.set_value ('Jhon',1000,'AT1000002')
```

```python
a2.display()
a1.display()
a1.check_balance()
```

```python
a1.withdraw(1000)
a1.display()
a1.deposit(4000)
a1.check_balance()
```

```python
class Shipment:
    company = "Zugvogel Speditions GmbH"
    total_shipments = 0

    def __init__(self, shipment_id, origin, destination, weight):
        self.shipment_id = shipment_id
        self.origin = origin
        self.destination = destination
        self.weight = weight
        self.status = "Pending"
        Shipment.total_shipments += 1

    @classmethod
    def get_company(cls):
        return cls.company

    @staticmethod
    def is_valid_weight(weight):
        return weight > 0

shp1 = Shipment("SHP001", "Mumbai", "Vienna", 120.5)
shp2 = Shipment("SHP002", "Delhi", "Berlin", 85.0)

print(Shipment.get_company())
print(Shipment.total_shipments)
print(Shipment.is_valid_weight(-10))
print(Shipment.is_valid_weight(50))
```
